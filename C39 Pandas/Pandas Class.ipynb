# Difference between unique and nunique

1.unique gives us the list of distinct values present in a column or list whereas nunique returns number of unique elements in the object. 

2. unique considers the NaN as well whereas NaN values are eliminated by nunique that is by default NaN values are not considered.

# Merge Function Python

```python
dict1={'Name':['Inderpreet','Shomendra','Sagar','Rahul','Rohit'],'Maths_Marks':[66,77,88,64,73],'Roll no':[1,2,3,4,1]}

df=pd.DataFrame(dict1)
df
```

```python
dict2={'Name':['Inderpreet','Shomendra','Sagar'],'Surname':['Singh','Sharma','Verma'],'Science_Marks':[70,94,98],'Roll no':[1,2,3]}

df2=pd.DataFrame(dict2)
df2
```

```python
merged=pd.merge(df,df2,on=['Roll no','Name'])
```

```python
merged
```

```python
merged['Full_name']=merged['Name']+" "+ merged['Surname']
merged
```

```python
merged.set_index('Roll no',inplace=True)
merged
```
